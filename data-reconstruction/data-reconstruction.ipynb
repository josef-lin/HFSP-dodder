```python
import numpy as np 
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt 
```

# Load file and visualise

```python
# Load file
df = pd.read_csv('synthetic_data_independent_views.csv')

# Inspect
print(df.head())
print(df.columns)
```

Single snapshot plot:

```python
# generation 1 time 5
g = 1
t = 5

subset = df[(df['gen'] == g) & (df['time'] == t)]

# split into projections 
xz = subset[subset['view'] == 'xz']
yz = subset[subset['view'] == 'yz']

# to numpy array 
xz = xz.sort_values('point')
yz = yz.sort_values('point')

x = xz['coord1'].values
z_xz = xz['coord2'].values

y = yz['coord1'].values
z_yz = yz['coord2'].values
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(x, z_xz, '-o')
plt.title('(x,z)')
plt.xlabel('x'); plt.ylabel('z')

plt.subplot(1,2,2)
plt.plot(y, z_yz, '-o')
plt.title('(y,z)')
plt.xlabel('y'); plt.ylabel('z')

plt.tight_layout()
plt.show()
```
